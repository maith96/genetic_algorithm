```python
import math

def softmax(scores):
    # Calculate the exponentials of the scores
    exp_scores = [math.exp(score) for score in scores]
    
    # Calculate the sum of the exponentials
    sum_exp_scores = sum(exp_scores)
    
    # Calculate the probabilities
    probabilities = [exp_score / sum_exp_scores for exp_score in exp_scores]
    
    return probabilities

# Example usage
scores = [2.0, 1.0, 0.1]
probabilities = softmax(scores)
print(f"Probabilities: {(probabilities)}")
for p in probabilities:
    print(f'{round(p, 1)*300}: {round(p, 1)}')
```

```python
import math

class Species:
    def __init__(self, name, population, performance_score):
        self.name = name
        self.population = population
        self.performance_score = performance_score

def softmax(scores):
    exp_scores = [math.exp(score) for score in scores]
    sum_exp_scores = sum(exp_scores)
    probabilities = [exp_score / sum_exp_scores for exp_score in exp_scores]
    return probabilities

def adjust_populations(species_list, total_population):
    # Extract performance scores from species list
    performance_scores = [species.performance_score for species in species_list]
    
    # Convert scores to probabilities
    probabilities = softmax(performance_scores)
    
    # Adjust population sizes based on probabilities
    for species, probability in zip(species_list, probabilities):
        species.population = round(probability * total_population)
    
    return species_list

# Example usage
species_list = [
    Species("Species A", 100, 2.0),
    Species("Species B", 100, 1.0),
    Species("Species C", 100, 0.1)
]

total_population = 300
adjusted_species_list = adjust_populations(species_list, total_population)

for species in adjusted_species_list:
    print(f"{species.name}: Population = {species.population}")

# Output will show the adjusted populations based on performance scores
```
